```python
import pandas as pd
path='data/'
filename=path+'Credit.csv'
credit = pd.read_csv(filename)
```

```python
credit.head()
```

```python
credit = credit[['Balance','Income','Limit','Rating','Cards','Age']]
credit.head()
```

```python
from pandas.plotting import scatter_matrix
%matplotlib inline
scatter_matrix(credit, alpha=0.2);
```

```python
# Stepwise

from sklearn.linear_model import LinearRegression
y = credit['Balance'].values
X = credit[['Income']].values
lr = LinearRegression()
lr.fit(X,y)
score1 = lr.score(X,y)

y = credit['Balance'].values
X = credit[['Limit']].values
lr = LinearRegression()
lr.fit(X,y)
score2 = lr.score(X,y)

y = credit['Balance'].values
X = credit[['Rating']].values
lr.fit(X,y)
score3 = lr.score(X,y)

print(score1,score2,score3)
```

```python
# Adding more variables

y = credit['Balance'].values
X = credit[['Rating', 'Income']].values
lr.fit(X,y)
score31 = lr.score(X,y)

y = credit['Balance'].values
X = credit[['Rating', 'Limit']].values
lr.fit(X,y)
score32 = lr.score(X,y)

print(score31,score32)

#Now we know that Income is our second best vriable to predict the value
```

```python
# Shrinkage ( Ridge & Lasso)

from sklearn.linear_model import Ridge
import numpy as np
y = credit['Balance'].values
X = credit[['Income','Limit','Rating','Cards','Age']]
# zero penalization gives the least squares
rr = Ridge(alpha=0,normalize=True)
rr.fit(X,y)
```

```python
X_pred = np.array([15,3000,300,2,34]).reshape(1,5)
rr.predict(X_pred)
```

```python
rr.coef_
```

```python
rr = Ridge(alpha=10,normalize=True)
rr.fit(X,y)
```

```python
rr.predict(X_pred)
```

```python
rr.coef_
```

```python
# Cross Validating
from sklearn.linear_model import RidgeCV
alpha_values = np.linspace(.0002,.02,num=100)
rrcv = RidgeCV(alphas=alpha_values,normalize=True, store_cv_values=True)
rrcv.fit(X,y)
```

```python
cv_values = np.sum(rrcv.cv_values_, axis=0)
import matplotlib.pyplot as plt
plt.plot(alpha_values, cv_values, 'or');
```

```python
print("Best regularization parameter:", rrcv.alpha_)
print("Coefficients are:",rrcv.coef_)
print("R squared: ",rrcv.score(X,y))
```

```python
# Now with Lasso

from sklearn.linear_model import Lasso
lr = Lasso(alpha=.1, normalize=True)
lr.fit(X,y)
lr.predict(X_pred)
lr.coef_
```

```python
from sklearn.linear_model import LassoCV
alpha_values = np.linspace(.0002,.02,num=100)
lrcv = LassoCV(alphas=alpha_values,normalize=True)
lrcv.fit(X,y)
```
